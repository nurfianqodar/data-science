```python
import pandas as pd
```

```python
df = pd.read_csv("./clean_data_cuaca_harian_kabupaten_bogor.csv")
df.sample(5)
```

```python
df["Tanggal"] = pd.to_datetime(df["Tanggal"])
df.sample(5)
```

```python
df["Month"] = df["Tanggal"].dt.month
df.sample(5)
```

```python
df_month = df.drop(columns=["Tanggal", "ddd_car"]).groupby(["Month"]).mean()
```

```python
df_month
```

```python
df_month = df_month[["Tn", "Tx", "Tavg", "RH_avg", "RR","ss"]]
df_month.columns = ["t_min", "t_max", "t_avg", "h_avg", "rr","ss"]
df_month
```

```python
df_month = df_month.round(4)
df_month
```

```python
df_month.to_excel("clean_avg_month_10_years_kab_bogor.xlsx")
```
